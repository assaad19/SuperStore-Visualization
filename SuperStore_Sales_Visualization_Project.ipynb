# SuperStore Visualization Project
In this project, I will visualize and analyze the superstore's dataset to gain a better understanding of the superstore's sales trends, correlations between attributes, its operations, customer segments, and product segments in order to tailor the best marketing strategy and advertising campaigns for the store.

## Superstore Dataset Overview
In my Superstore dataset, which I found on Kaggle, I have a variety of attributes, including 6 numerical, 8 categorical, 5 string, and 2 date attributes, with a total of 9994 rows. Here is the list of attributes:

- **Row ID**: ID of each Row (Numerical)
- **Order ID**: Unique order number (String)
- **Order Date**: The date the order was placed (Date)
- **Ship Date**: The date the order was shipped (Date)
- **Ship Mode**: The type of shipment for each order (Categorical)
- **Customer Name**: The name of each customer (String)
- **Customer ID**: The unique ID of each customer (String)
- **Segment**: Customer segmentation category (Categorical)
- **Country**: The destination country of each order (Categorical)
- **City**: The destination city of each order (Categorical)
- **State**: The destination state of each order (Categorical)
- **Postal Code**: The postal code of each order (Numerical)
- **Region**: The destination region of each order (Categorical)
- **Category**: Generalized category of the products (Categorical)
- **Sub-Category**: Detailed category of the products (Categorical)
- **Product Name**: The name of each product (String)
- **Product ID**: ID of each product (String)
- **Sales**: Revenue of each order (Numerical)
- **Quantity**: Quantity sold in each order (Numerical)
- **Discount**: Discount applied to each order (Numerical)
- **Profit**: Profit made on each order (Numerical)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

## Reading the Dataset

Read the data from a CSV file into a pandas DataFrame.

```python
df=pd.read_csv(r"/content/Sample - Superstore.csv",encoding="latin1")
print(df.head())
```

## Data Preprocessing

Clean and prepare the data for analysis,removing unwanted columns and converting some date attributs to make the work easier

```python
#Removing ID columns

df.drop(columns=['Row ID','Order ID','Customer ID','Product ID'],inplace=True)
print(df.head())

#converting some date attributs to make the work easier

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
print(df.head())
print(df.info())
```

###Pie Chart for Category Distribution and Segments


```python
plt.subplot(1,2,1)
category=df['Category'].value_counts().sort_index()
plt.pie(category,autopct='%1.1f%%')
plt.title("Category Distribution")
plt.legend(df['Category'].unique(), title='Category',fontsize=6.5 ,bbox_to_anchor=(1, 0))

plt.subplot(1,2,2)
segments=df['Segment'].value_counts().sort_index()
plt.pie(segments,autopct='%1.1f%%')
plt.title('Segements Distribution')
plt.legend(df['Segment'].unique(), title='Segments',fontsize=6.5 ,bbox_to_anchor=(1, 0))
plt.show()
```

To visualize the distribution of categories and segments, I created two pie charts using `plt.pie()` and arranged them side by side using `plt.subplot()`. The results show that the best-selling category is office supplies, representing 60.3% of the total. Furthermore, the Superstore appears to be more oriented towards B2C (business-to-consumer) commerce rather than B2B (business-to-business), as consumers represent 51.9% of the segment. We have also discovered something interesting: there may be a correlation between the segment type and the category, as the percentages are similar (21-30 between corporate and furniture, 60-52 between office supplies and consumers, 18-18.5 between technology and home office). Now, I will verify if this is true by examining which categories are most purchased by each segment among the three available.



###Radar Chart for category and segments


```python
# Define colors for each segment
colors = {'Consumer': 'blue', 'Home Office': 'green', 'Corporate': 'orange'}

# Create subplots for each segment
fig, ax = plt.subplots(figsize=(5, 5), subplot_kw=dict(polar=True)) #

for segment, color in colors.items():
    # Filter the data for the current segment
    segment_df = df[df["Segment"] == segment]

    # Select the columns for the products and calculate the count by category
    products = segment_df["Category"].value_counts().sort_index()

    # Create a radar chart for the count of product consumption
    theta = np.linspace(0, 2*np.pi, len(products), endpoint=False)
    ax.plot(theta, products, color=color, marker='o', label=segment)  # Use specified color and add markers
    ax.fill(theta, products, color=color, alpha=0.3)  # Fill the radar area with alpha for transparency

# Add labels and a legend to the plot
ax.set_thetagrids(np.degrees(theta), products.index)
ax.set_title("Distribution of Product Categories by Segment")
ax.legend(loc='upper right')

# Show the plot
plt.show()
```

To test the hypothesis above and see how categories are distributed across segments, I created a radar chart by iterating through the `Colors` variable with the name of each segment and its corresponding color. Then, I filtered my segments using `segment_df = df[df["Segment"] == segment]`, where `segment` is the index of the color variable. Next, I counted the product category for each segment using `segment_df["Category"].value_counts().sort_index()`. I created the circles for the radar chart using `np.linspace(0, 2*np.pi, len(products), endpoint=False)`, and then I plotted and filled each one using `ax.plot()` and `ax.fill()`. Finally, `ax.set_thetagrids(np.degrees(theta), products.index)` sets the theta grid labels based on the degree values and product categories.

In conclusion, we can determine that there is no correlation between them, and the best-selling category across all segments is office supplies, which is well-known at SuperStore.


###Scatter plot for sales and profits by category


```python
plt.figure(figsize=(5, 5))
colors = {'Furniture': 'royalblue', 'Office Supplies': 'orangered', 'Technology': 'forestgreen'}
markers = {'Furniture': 'o', 'Office Supplies': 's', 'Technology': '^'}
for category, color in colors.items():  # Define a color palette for different categories
    subset = df[df['Category'] == category]
    plt.scatter(subset['Sales'], subset['Profit'], label=f'{category}', color=color, marker=markers[category], alpha=0.5)
plt.xlabel('Sales')
plt.ylabel('Profits')
plt.title('Scatter Plot of Sales and Profits by Category')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

To visualize the distribution between sales and profits by category, I used `plt.scatter()`. The results show that we have significant losses in office supplies and furniture, and the data points are very close to each other. I concluded that they might have a similar profit margin for their sales. To test this hypothesis, I created a bar chart to visualize the numbers more clearly.



###BarChart for profits and sales by category


```python
category_data = df.groupby('Category')[['Sales', 'Profit']].sum()
plt.figure(figsize=(8, 8))
bar_width = 0.35
x = range(len(category_data))
colors = ['Blue', 'Orange'] #Create the bar chart with good color choices
plt.bar(x, category_data['Sales'], width=bar_width, label='Sales', color=colors[0], align='center')
plt.bar([i + bar_width for i in x], category_data['Profit'], width=bar_width, label='Profit', color=colors[1], align='center')
plt.xticks([i + bar_width / 2 for i in x], category_data.index)
for i in x:     #adding text for the amount
    plt.text(i, category_data['Sales'][i], f"${category_data['Sales'][i]:.2f}", ha='center', va='bottom')
for i in x:
    plt.text(i + bar_width, category_data['Profit'][i], f"${category_data['Profit'][i]:.2f}", ha='center', va='bottom')
plt.title("Total Sales and Profit by Category")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

To understand why the data points are so close to each other, I created a bar chart using `plt.bar()`. It turned out that the profit margins are indeed the same, with 16% for Office Supplies and Technology, and less than 5% for Furniture. This confirms the hypothesis we observed in the scatter plot. However, to have a more comprehensive view, I now need to examine the total profits and losses for this category.


###BarChart for profits by subcategories


```python
subcategory_profits = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
subcategory_profits.plot(kind='bar', color='skyblue')
for i, v in enumerate(subcategory_profits):
    plt.text(i, v + 50, f"${v:.2f}", ha='center', va='bottom',fontsize=8)
plt.title("Total Profits by Subcategory")
plt.xlabel("Subcategory")
plt.ylabel("Total Profits")
plt.xticks(rotation=45)
plt.show()
```

In this chart, we can observe that the largest loss of $17,725$ is associated with tables, which belong to the Furniture category, ultimately confirming our hypothesis. The highest profit generated comes from photocopiers, amounting to $55,617$

###Scatter of Profits by each month from 2014 till 2017


```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Order Month"] = df["Order Date"].dt.month
df["Order Year"] = df["Order Date"].dt.year

profit_sum_by_month = df.groupby(["Order Year", "Order Month"])["Profit"].sum().reset_index()

#Each year points:
#2014:
profit_sum_2014 = profit_sum_by_month.loc[profit_sum_by_month['Order Year']==2014]
#2015
profit_sum_2015 = profit_sum_by_month.loc[profit_sum_by_month['Order Year']==2015]
#2016
profit_sum_2016 = profit_sum_by_month.loc[profit_sum_by_month['Order Year']==2016]
#2017
profit_sum_2017 = profit_sum_by_month.loc[profit_sum_by_month['Order Year']==2017]

plt.rcParams.update({'font.size': 10}) #to update font size of text for 10

plt.scatter(profit_sum_2014["Order Month"], profit_sum_2014["Profit"], label = "2014")
plt.scatter(profit_sum_2015["Order Month"], profit_sum_2015["Profit"], label = "2015")
plt.scatter(profit_sum_2016["Order Month"], profit_sum_2016["Profit"], label = "2016")
plt.scatter(profit_sum_2017["Order Month"], profit_sum_2017["Profit"], label = "2017")
plt.xticks(range(1, 13))
plt.yticks(range(-5000,20001,5000))
plt.ylim(-5000,20000)
plt.title("Profit by month in each year")
plt.legend()
plt.xlabel("Month")
plt.ylabel("Total Profit")
#Regression Line:
x=profit_sum_by_month["Order Month"]
y=profit_sum_by_month["Profit"]
z = np.polyfit(x, y, 1) #the 1 indicates that i am fitting a 1st degree linear polynomial(calculating the coefficients)
p = np.poly1d(z) #creating it
plt.plot(x, p(x), color="purple")
plt.show()
```

To clearly visualize the variation in profits for each month of each year, I created four scatter plots after grouping my data by year and month by extracting the month and year from the "Order Date" column using `df["Order Date"].dt.month`. I then separated the data by year using `loc` for each year. Next, I plotted a linear regression line using `z=np.polyfit(x, y, 1)` to fit the data and applied it to the data using `np.poly1d(z)` to see that there is a positive relationship in profits between the months of 2014 and 2017. Additionally, in December 2016, they achieved the highest profit, and in January 2015, they experienced the greatest loss. I also used `plt.rcParmsa.update({'font.size' : 10})` to update the text size to 10, where `rcParams` is a function containing many parameters for customizing the appearance of plots.

###Order per day of week


```python
df['order_date_day'] = df['Order Date'].dt.day
df['order_date_dayofweek'] = df['Order Date'].dt.dayofweek
df['order_date_dayofweek_name'] = df['Order Date'].dt.strftime('%a')
print(df['order_date_dayofweek'].head())
plt.figure(figsize=(11, 11))
weekday_label = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
ax=sns.countplot(df, x='order_date_dayofweek_name',order=weekday_label,palette='YlGnBu')

total = len(df)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}\n {height/total:.2%}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom') # to put the text at top of the bar in the middle

plt.xlabel('Day of Week')
plt.ylabel('Total Orders')
plt.ylim(0,2000)
plt.title('Total Orders by Day of Week', size=14, color='dimgrey', pad=20)
plt.show()
```

##Conclusion:
After visualizing the SuperStore data, it is clear that they are performing well each year for a store of this size, and their growth is healthy. However, if I were to provide a recommendation to the owner on how to increase sales, the focus should be on the following:

1. Target Consumer Segment: Concentrate efforts on targeting customers from the consumer segment, as they accounted for the highest sales and quantity of items ordered across all segments.

2. Staples-related Products: Based on the Exploratory Data Analysis (EDA), prioritize the sale of products related to office supplies (e.g., Staples), as most customers have ordered them in superstores.

3. Marketing and Promotion Timing: Allocate advertising spending primarily to November, December, and September, as these months have shown the highest sales. Focusing promotional efforts during these periods can be more effective.

4. Day of the Week: Consider running advertisements specifically on Mondays since we observed that Mondays have the highest order volume. This can make marketing promotions more effective by reaching potential customers at the ideal time.

5. Geographic Focus: Concentrate marketing efforts on the Western region of the United States. For cities, prioritize advertising in New York, Los Angeles, Philadelphia, and San Francisco. For states, target California, New York, and Texas for marketing promotions.

By following these recommendations, the SuperStore can further improve its sales and continue its healthy growth trajectory.
